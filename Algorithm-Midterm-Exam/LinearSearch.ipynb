```python
# Linear Search
import random
import time
import timeit

start = time.time()

def linearSearch(arr, x):
 
    for i in range(len(arr)):
        if (arr[i] == x):
            return i
    return -1

n_values = range(10, 1010, 10)
linear_times = []

# Loop over different n values
for n in n_values:

    # Initialize list S with random integers
    S = [random.randint(1, 1000) for _ in range(n)]

    # 選擇 integer x
    x = random.randint(1, 1000)

    # 執行演算法5次，並計算平均執行時間
    linear_time = timeit.timeit(lambda: linearSearch(S, x), number = 5) / 5

    # 將每一次的平均執行時間 Append 到 list 中
    linear_times.append(linear_time)

    print("第", len(linear_times), "次，S = ", len(linear_times) * 10, "時，執行的時間為", linear_times[(n // 10) - 1])

# 結束測量
end = time.time()

# 輸出結果
print("總執行時間：%f 秒" % (end - start))
```
