```python
import random
import time
import timeit

start = time.time()

def binarySearch(arr, x):
    low = 0
    high = len(arr) - 1    

    while low <= high:
 
        mid = (low + high) // 2
 
        if arr[mid] == x:
            return mid
 
        elif arr[mid] < x:
            low = mid + 1
 
        else:
            high = mid - 1

    return -1

n_values = range(10, 1010, 10)
binary_times = []

# Loop over different n values
for n in n_values:

    # Initialize list S with random integers
    S = [random.randint(1, 1000) for _ in range(n)]

    # 選擇 integer x
    x = random.randint(1, 1000)

    # 執行演算法5次，並計算平均執行時間
    binary_time = timeit.timeit(lambda: binarySearch(S, x), number = 5) / 5

    # 將每一次的平均執行時間 Append 到 list 中
    binary_times.append(binary_time)

    try:
        index = S.index(x)
        print("第", len(binary_times), "次，S =", len(binary_times)*10, "時，執行的時間為", binary_times[(n // 10) - 1],"，Element found in index:",index)
    except ValueError:
        print("第", len(binary_times), "次，S =", len(binary_times)*10, "時，執行的時間為", binary_times[(n // 10) - 1],"，Element not found in the list")

# 結束測量
end = time.time()

# 輸出結果
print("總執行時間：%f 秒" % (end - start))
```
